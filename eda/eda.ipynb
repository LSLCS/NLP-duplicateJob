```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "../data/jobs.csv"  # Update this if needed
df = pd.read_csv(file_path)
```

```python
# Display basic information
print("Dataset Overview:")
display(df.head())
```

```python
# Summary statistics
print("\nDataset Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Visualize missing values
plt.figure(figsize=(8, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df["jobTitle"].value_counts().head(20)
```

```python
# Distribution of job titles
plt.figure(figsize=(10, 5))
ax = df["jobTitle"].value_counts().head(20).plot(kind="bar")
plt.title("Top 20 Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=75)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=10)

plt.show()
```

```python
# Top companies with most job postings
plt.figure(figsize=(12, 6))
df["companyName"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Companies with Job Postings")
plt.xlabel("Company")
plt.ylabel("Number of Postings")
plt.xticks(rotation=45)
plt.show()
```

```python
# Length of job descriptions
df["description_length"] = df["jobDescRaw"].astype(str).apply(len)
plt.figure(figsize=(8, 5))
sns.histplot(df["description_length"], bins=50, kde=True)
plt.title("Distribution of Job Description Lengths")
plt.xlabel("Character Count")
plt.ylabel("Frequency")
plt.show()
```

```python
# Show dataset insights
print("\nUnique Job Titles:", df["jobTitle"].nunique())
print("Unique Companies:", df["companyName"].nunique())
```

```python
# Check for duplicate job descriptions
duplicates = df.duplicated(subset=["jobDescRaw"], keep=False).sum()
print(f"\nPotential duplicate job postings: {duplicates}")
```

```python
# Check for top job titles with duplicate job descriptions
df['is_duplicate'] = df.duplicated(subset=["jobDescRaw"], keep=False)

duplicated_jobs = df[df['is_duplicate'] == True]

# Count the number of duplicated postings for each company
titles_dup_count = duplicated_jobs['jobTitle'].value_counts()

# Get the top N companies with the most duplicated job postings
top_titles = titles_dup_count.head(10)

# Plot the distribution of duplicated job postings by company
plt.figure(figsize=(10, 6))
sns.barplot(x=top_titles.index, y=top_titles.values, palette="viridis")

# Add labels and title
plt.title("Top Job Titles with Duplicated Job Postings")
plt.xlabel("Job titles")
plt.ylabel("Count of Duplicated Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Check for top companies with duplicate job descriptions
df['is_duplicate'] = df.duplicated(subset=["jobDescRaw"], keep=False)

duplicated_jobs = df[df['is_duplicate'] == True]

# Count the number of duplicated postings for each company
company_dup_count = duplicated_jobs['companyName'].value_counts()

# Get the top N companies with the most duplicated job postings
top_companies = company_dup_count.head(10)

# Plot the distribution of duplicated job postings by company
plt.figure(figsize=(10, 6))
sns.barplot(x=top_companies.index, y=top_companies.values, palette="viridis")

# Add labels and title
plt.title("Top Companies with Duplicated Job Postings")
plt.xlabel("Company")
plt.ylabel("Count of Duplicated Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Show the top job descriptions for the company with highest duplicated job description
# Filter for rows where the company is Thriveworks
thriveworks_jobs = df[df['companyName'] == "Thriveworks"]

# Filter for duplicated job descriptions
thriveworks_duplicated_jobs = thriveworks_jobs[thriveworks_jobs['is_duplicate'] == True]

# Count the occurrences of each duplicated job description
top_duplicated_jobs = thriveworks_duplicated_jobs['jobTitle'].value_counts()

# Get the top N duplicated job descriptions (for example, top 10)
top_duplicated_jobs = top_duplicated_jobs.head(10)

# Display the top duplicated job descriptions
print(top_duplicated_jobs)

# Create a bar plot for the job titles with the top duplicated job descriptions
plt.figure(figsize=(10, 6))
sns.barplot(x=top_duplicated_jobs.values, y=top_duplicated_jobs.index, palette="viridis")

# Add labels and title
plt.title("Job titles for Top Duplicated Job Descriptions for Thriveworks")
plt.xlabel("Count of Duplicated Job Descriptions")
plt.ylabel("Job Titles")
plt.show()
```

```python
# Show the top job descriptions for the company with second highest duplicated job description
# Filter for rows where the company is Thriveworks
thriveworks_jobs = df[df['companyName'] == "LocumJobsOnline"]

# Filter for duplicated job descriptions
thriveworks_duplicated_jobs = thriveworks_jobs[thriveworks_jobs['is_duplicate'] == True]

# Count the occurrences of each duplicated job description
top_duplicated_jobs = thriveworks_duplicated_jobs['jobTitle'].value_counts()

# Get the top N duplicated job descriptions (for example, top 10)
top_duplicated_jobs = top_duplicated_jobs.head(10)

# Display the top duplicated job descriptions
print(top_duplicated_jobs)

# Create a bar plot for the job titles with the top duplicated job descriptions
plt.figure(figsize=(10, 6))
sns.barplot(x=top_duplicated_jobs.values.astype(int), y=top_duplicated_jobs.index, palette="viridis")

# Add labels and title
plt.title("Job titles for Top Duplicated Job Descriptions for LocumJobsOnline")
plt.xlabel("Count of Duplicated Job Descriptions")
plt.ylabel("Job Titles")
unique_ticks = sorted(set([int(i) for i in plt.gca().get_xticks()]))
plt.xticks(ticks=unique_ticks, labels=unique_ticks)
plt.show()
```
